The last twenty documents uploaded through Magdatum:

```sql
SELECT TOP 20 *
  FROM [dbo].[DocumentUploadHistory]
  ORDER BY [Datetime] DESC
```

The Magdatum Event Log sorted by latest to earliest:

```sql
SELECT *
  FROM [dbo].[Logs]
  ORDER BY [Timestamp] DESC
```

The Matter list used by Magdatum's uploader:

```sql
SELECT *
FROM [dbo].[Matters]
ORDER BY [MatterNumber] ASC
```

The Matters as downloaded from LEAP

```sql
SELECT *
FROM [dbo].[MattersLeap]
ORDER BY [FileNumber] ASC
```

The Matters as downloaded from Lexis Nexis Affinity

```sql
SELECT *
FROM [dbo].[MattersAffinity]
ORDER BY [MatterId] ASC
```
